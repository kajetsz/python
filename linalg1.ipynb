```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as ln
import scipy.optimize as op
```

```python
#ex1

A = np.array([[2, 3, 4], [5, 4, 23]])

plt.figure()

x = np.linspace(-100, 100, 1000)

y1 = (4 - 2*x) / 3
plt.plot(x, y1)

y2 = (23 - 5*x) / 4
plt.plot(x, y2)

solution = np.linalg.solve(A[:,0:2], A[:,2])

plt.scatter(solution[0], solution[1])

#LU decomposition

P, L, U = ln.lu(A)
print(P,'\n', L,'\n', U)


plt.show()


rank  = np.linalg.matrix_rank(A)
condition = np.linalg.cond(A)

print('Rank:', rank, '\nConditioning number:', condition)
```
